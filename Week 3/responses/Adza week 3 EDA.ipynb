## Data Cleaning

```python
# Convert pickup and dropoff datetime columns
nyc_taxi_trip_duration['pickup_datetime'] = pd.to_datetime(nyc_taxi_trip_duration['pickup_datetime'])
nyc_taxi_trip_duration['dropoff_datetime'] = pd.to_datetime(nyc_taxi_trip_duration['dropoff_datetime'])
```

```python
# Create trip duration in minutes (if not already present)
if 'trip_duration' not in nyc_taxi_trip_duration.columns:
    nyc_taxi_trip_duration['trip_duration'] = (nyc_taxi_trip_duration['dropoff_datetime'] - nyc_taxi_trip_duration['pickup_datetime']).dt.total_seconds() / 60
```

```python
# Remove trips with 0 or negative duration
nyc_taxi_trip_duration = nyc_taxi_trip_duration[nyc_taxi_trip_duration['trip_duration'] > 0]
```

```python
# Log transformation if skewed
nyc_taxi_trip_duration['log_trip_duration'] = np.log1p(nyc_taxi_trip_duration['trip_duration'])

sns.histplot(nyc_taxi_trip_duration['log_trip_duration'], bins=100, kde=True)
plt.title("Log-Transformed Trip Duration")
plt.xlabel("log(Duration)")
plt.show()
```

## Datetime Feature Engineering

```python
# Convert to datetime
nyc_taxi_trip_duration['pickup_datetime'] = pd.to_datetime(nyc_taxi_trip_duration['pickup_datetime'])
nyc_taxi_trip_duration['dropoff_datetime'] = pd.to_datetime(nyc_taxi_trip_duration['dropoff_datetime'])

# Extract time features
nyc_taxi_trip_duration['pickup_hour'] = nyc_taxi_trip_duration['pickup_datetime'].dt.hour
nyc_taxi_trip_duration['pickup_day'] = nyc_taxi_trip_duration['pickup_datetime'].dt.day
nyc_taxi_trip_duration['pickup_weekday'] = nyc_taxi_trip_duration['pickup_datetime'].dt.weekday
```

## Trip Duration by Time

```python
# Trip count by hour
sns.countplot(x='pickup_hour', data=nyc_taxi_trip_duration)
plt.title("Trips by Hour of Day")
plt.xlabel("Hour")
plt.ylabel("Trip Count")
plt.show()
```

```python
# Trip duration by hour of day
sns.boxplot(x='pickup_hour', y='trip_duration', data=nyc_taxi_trip_duration)
plt.title("Trip Duration by Pickup Hour")
plt.ylim(0, 5000)
plt.show()
```

```python
# Trip duration by weekday
sns.boxplot(x='pickup_weekday', y='trip_duration', data=nyc_taxi_trip_duration)
plt.title("Trip Duration by Day of Week")
plt.ylim(0, 5000)
plt.show()
```

## Geospatial Visualization

```python
# Scatter plot of pickup locations (sample for performance)
sample_df = nyc_taxi_trip_duration.sample(10000)

# Pickup heat scatter
plt.scatter(sample_df['pickup_longitude'], sample_df['pickup_latitude'], alpha=0.3, s=1)
plt.title("Pickup Location Density")
plt.xlabel("Longitude")
plt.ylabel("Latitude")
plt.show()

# Dropoff heat scatter
plt.scatter(sample_df['dropoff_longitude'], sample_df['dropoff_latitude'], alpha=0.3, s=1)
plt.title("Dropoff Location Density")
plt.xlabel("Longitude")
plt.ylabel("Latitude")
plt.show()
```
